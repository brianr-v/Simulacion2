# Distancia Manhattan

## Caminta en un plano

Iniciamos suponiendo que una persona que se puede mover al azar en cuatro direcciones (N, S, E, O). Despues de varios pasos, la persona se encontrara en una posición final $(x_f,y_f)$, entonces queremos determinar 

$\begin{equation}d_m=|x_f - x_i| + |y_f-y_i|\end{equation}$

Como ejercicio, necesitamos determinar, dado N pasos encontrar a una longuitud no más de 4 del 50%.

# Simulación 

## Importación de las librerías

```python
import random
import matplotlib.pyplot as plt
```

## Definimos funciones

* Función movimiento():
Donde definimos las variables "x" y "y", lo que hace es generar un numero aleatorio que serán los pasos que dara dependiendo (Norte, Sur, Este, Oeste). Reegresa "x" y "y", que sera la posición final.

* Función probabilidad(n, k):
Donde calculo la distancia Manhattan con la formula:
$\begin{equation}d_m=|x_f - x_i| + |y_f-y_i|\end{equation}$
Regresa la media de las distancias.

```python
def movimiento(n):
    y = 0
    x = 0
    for i in range(n):
        aleatorio = random.randint(0,3)
        if aleatorio == 0:
            y += 1
        elif aleatorio == 1:
            y -= 1
        elif aleatorio == 2:
            x += 1
        else:
            x -= 1
    return x, y

def probabilidad(n,k):
    contador = 0

    for i in range(n):
        xi = 0
        yi = 0
        sim = movimiento(k)
        xf = sim[0]
        yf = sim[1]
        
        if (abs(xi - xf) + abs(yi-yf)) <= 4:
            contador += 1
    return contador/n
```

## Llamamos a las funciones

```python
lineas = 100
n = 20000

lista_probabilidad = []

for i in range(1,lineas+1):
    
    proba = probabilidad(n,i)
    lista_probabilidad.append(proba)
    print("Probabilidad: ",proba, " con ", i, " caminos")

numero_mas_cercano = min(lista_probabilidad, key=lambda x: abs(x - 0.5))
print("\n El número mas cercano a 0.5 es, ",numero_mas_cercano, " que es el camino ", lista_probabilidad.index(numero_mas_cercano) + 1)
```

## Definimos las funciones

* Función probabilidad(n, k):
Donde calculo la distancia Manhattan con la formula:
$\begin{equation}d_m=|x_f - x_i|\end{equation}$
Regresa la media de las distancias.

```python
def movimiento(n):
    x = 0
    for i in range(n):
        aleatorio = random.randint(0,1)
        
        if aleatorio == 0:
            x += 1
        else:
            x -= 1
    return x

def probabilidad(n,k):
    contador = 0

    for i in range(n):
        xi = 0
        sim = movimiento(k)
        xf = sim
        
        if (abs(xi - xf) ) <= 4:
            contador += 1
    return contador/n
```
